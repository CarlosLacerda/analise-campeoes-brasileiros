```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("campoes_brasileiro.csv")

# Contar títulos por time
titulos = df['Vencedor'].value_counts()
print(titulos)

# Fazer um gráfico de barras
plt.figure(figsize=(12, 6))
titulos.plot(kind='bar', color='green')
plt.title('Títulos do Campeonato Brasileiro por Time')
plt.xlabel('Time')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

timeline = df[['Ano', 'Vencedor']].copy()

plt.figure(figsize=(15, 8))
times_unicos = df['Vencedor'].unique()
cores = plt.cm.tab20(range(len(times_unicos)))
cor_map = dict(zip(times_unicos, cores))

for time in times_unicos:
    dados_time = timeline[timeline['Vencedor'] == time]
    plt.scatter(dados_time['Ano'], [time]*len(dados_time), 
                s=200, c=[cor_map[time]], label=time, alpha=0.7)

plt.xlabel('Ano')
plt.ylabel('Campeão')
plt.title('Timeline de Campeões Brasileiros')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
